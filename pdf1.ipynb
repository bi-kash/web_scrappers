```python
pdf_path = 'GUILFORD-20161108-Style050.pdf'

import math
from collections import Counter
def calculate_center(x0, y0, x1, y1):
    """Calculate the center of a bounding box."""
    center_x = (x0 + x1) / 2
    center_y = (y0 + y1) / 2
    return center_x, center_y

def calculate_distance(center1, center2):
    """Calculate the Euclidean distance between two points."""
    return math.sqrt((center1[0] - center2[0]) ** 2 + (center1[1] - center2[1]) ** 2)

def inches_to_points(inches):
    """Convert inches to points."""
    return [int(i * 72) for i in inches]

def get_mode_color(image, bbox):
    """Get the mode (most frequent) color of the region defined by bbox in the image."""
    cropped_image = image.crop(bbox)
    pixels = list(cropped_image.getdata())
    mode_color = Counter(pixels).most_common(1)[0][0]
    return mode_color
```

```python
import re

def divide_names(name_string):
    # Split the string by spaces
    words = name_string.split()
    
    # Assuming that names can have either two or three parts (first, last, or first, middle, last)
    divided_names = []
    temp_name = []
    
    for word in words:
        temp_name.append(word)
        if len(temp_name) == 2 or len(temp_name) == 3:
            divided_names.append(" ".join(temp_name))
            temp_name = []
    
    # If any leftover name part exists (edge case)
    if temp_name:
        divided_names.append(" ".join(temp_name))
    
    return divided_names

names_string = "John Doe Jane Mary Johnson Alice Cooper Bob Steven Smith"
names_list = divide_names(names_string)
print(names_list)
```
